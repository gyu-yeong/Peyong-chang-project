```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
```

```python
latitude = []
longitude =[]

for i in df['행정구역']:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
df['경도'] = latitude
df['위도'] = longitude
df
```

#### 등분산 가정 위배

```python
from scipy.stats import mannwhitneyu

def mann_whitney_u_test(data1, data2):
    stat, p = mannwhitneyu(data1, data2)
    alpha = 0.05
    
    print('statistic:',stat)
    print('p-value:',p)
    print('-'*50)
    
    if p > alpha:
        print("fail to reject H0")
    else:
        print("reject H0")

mann_whitney_u_test(south_d,north_d)        
```

```python
from scipy import stats

t_statistic, p_value  = stats.ttest_ind(south_d, north_d)

print('statistic:',t_statistic)
print('p_value:',p_value)
print('-'*50)
if p_value > 0.05:
    print("남부와 북부의 평균소득에는 차이가 없다")
else:
    print("남부와 북부의 평균소득에는 차이가 있다.")
```
